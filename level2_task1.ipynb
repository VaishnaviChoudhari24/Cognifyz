Restuarnats ratings

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/Dataset .csv')
df.head()
```

Analyze distribution of aggregate ratings and determine most common rating range

```python
print(df.columns)
```

```python
agg_rating_distribution=df['Aggregate rating'].value_counts()
print(agg_rating_distribution)
```

```python
a=agg_rating_distribution.idxmax()
print("most commmon rating range is: ",a)
```

```python
plt.hist(df['Aggregate rating'],bins=30,color='skyblue',edgecolor='black')
plt.xlabel('Aggregate Rating')
plt.ylabel('Counts')
plt.title('Distribution of Aggregate Ratings')
plt.show()
```

Calculate average number of votes recieved by restuarants

```python
avg_vote=round(df['Votes'].mean(),3)
print("average number of votes recieved by restuarants: ",avg_vote)
```
